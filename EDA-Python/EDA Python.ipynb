## Maximum Likelihood Estimation (MLE)

In simple cases the MLE is
available in closed form. For example, if observations were iid from N(mu; sigma^2), then we already
know that the MLE for mu and sigma^2 are mu.hat = X_bar and sigma.hat^2 = (n-1)S^2/n. Thus, the ML estimates
for the extreme rainfall data can be quickly computed as

```python
mu_hat = np.mean(s1)
mu_hat
```

```python
n = len(s1)
sigma_hat = np.sqrt((n-1)*np.var(s1, ddof = 1)/n)
sigma_hat
```

```python
from scipy.optimize import minimize 
```

```python
def MLE_Gumb(param):
    loc, scale = param
    
    # Calculate the log-likelihood for gumbel distribution
    LL = np.sum(stats.gumbel_r.logpdf(s1, loc, scale))
    # Calculate the negative log-likelihood
    neg_LL = -1*LL
    return neg_LL 
```

```python
# minimize arguments: function, intial_guess_of_parameters, method
mle_model = minimize(MLE_Gumb, np.array([50, 10]), method='L-BFGS-B')
mle_model 
```

```python
def MLE_Norm(param):
    loc, scale = param
    
    # Calculate the log-likelihood for gumbel distribution
    LL = np.sum(stats.gumbel_r.logpdf(s1, loc, scale))
    # Calculate the negative log-likelihood
    neg_LL = -1*LL
    return neg_LL 
```
